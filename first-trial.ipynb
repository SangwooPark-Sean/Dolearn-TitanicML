```python
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier #성능평가용
import numpy as np
import pandas as pd
```

train data와 train target 으로 훈련한 모델로 test data 예측하여 test prediction 제출하기

```python
# train data 가공하기 : 사용될 features = Pclass, Sex, Age, Sibsp, Parch
train_data = pd.read_csv("/kaggle/input/titanic/train.csv")
train_data.drop(['PassengerId', 'Survived', 'Name', "Cabin", 'Embarked', 'Ticket', 'Fare'], axis = 1, inplace = True)

# 비어있는 Age값을 중간값으로 채워넣기
train_data['Age'] = train_data['Age'].fillna(train_data['Age'].median())

#male을 1로, female 을 2로 변환하기
for i in range(train_data.shape[0]):
    if train_data['Sex'][i] == 'male':
        train_data['Sex'][i] = 1
    else:
        train_data['Sex'][i] = 2

train_data.shape
```

```python
# train data의 Survived 값
train_result = pd.read_csv("/kaggle/input/titanic/train.csv").Survived
train_result.shape
```

```python
# test data 가공하기
test_data = pd.read_csv("/kaggle/input/titanic/test.csv")
test_data.drop(['PassengerId', 'Name', "Cabin", 'Embarked', 'Ticket', 'Fare'], axis = 1, inplace = True)

# 비어있는 Age값을 중간값으로 채워넣기
test_data['Age'] = test_data['Age'].fillna(test_data['Age'].median())

#male을 1로, female 을 2로 변환하기
for i in range(test_data.shape[0]):
    if test_data['Sex'][i] == 'male':
        test_data['Sex'][i] = 1
    else:
        test_data['Sex'][i] = 2

test_data.shape
```

```python
# Randomforest 모델생성
rfc = RandomForestClassifier(n_estimators=10)
rfc.fit(train_data, train_result)
```

```python
# 예측
prediction = rfc.predict(test_data)
```

```python
for_id = pd.read_csv("/kaggle/input/titanic/test.csv")
pd.DataFrame(data = {'PassengerId': for_id.PassengerId,
                     'Survived': prediction.astype(int)}).to_csv('first_trial_submission.csv', index = False)
```
